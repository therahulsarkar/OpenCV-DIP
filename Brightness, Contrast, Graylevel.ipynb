# Brightness enhance 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

zero_img = np.zeros_like(img)

final = cv2.addWeighted(img, 1.5, zero_img, 0, 0)

result = np.hstack((img, final))
plt.imshow(result)
```

# Brightness enhance Method:2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread("./img/elon.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

row,col =image.shape

k=70

for i in range(row):
    for j in range(col):
        a=image[i][j]+k
        if(a > 255):
            image[i][j]=255
        else:
            image[i][j]=a

plt.imshow(image, cmap="gray")
```

# Bightness supress Method: 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread("./img/elon.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

row,col =image.shape

k=70

for i in range(row):
    for j in range(col):
        a=abs(image[i][j]-k)
        if(a < 0):
            image[i][j]=0
        else:
            image[i][j]=a
plt.imshow(image, cmap="gray")
```

# Low contrast 

```python
import cv2
import numpy as np
image=cv2.imread("./img/elon.jpg",0)
row,col=image.shape
k=12

for i in range(row):
    for j in range(col):
        image[i][j] = round((image[i][j]/k),0)
        
        
plt.imshow(image, cmap="gray")
```

# High contrast 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread("./img/elon.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

row,col =image.shape

k=2

for i in range(row):
    for j in range(col):
        a=image[i][j]*k
        if(a > 255):
            image[i][j]=255
        else:
            image[i][j]=a

plt.imshow(image, cmap="gray")
```

# Gray level slicing 

```python
#gray level slicing without background
import cv2
import numpy as np
image=cv2.imread('./img/elon.jpg',0)

a=int(input('Enter the starting range:'))
b=int(input('Enter the end range:'))

h,b=image.shape

for i in range(h):
    for j in range(b):
        if(image[i][j]>=a and image[i][j]<=b):
            image[i][j]=255
        else:
            image[i][j]=0
            
plt.imshow(image, cmap="gray")
```
