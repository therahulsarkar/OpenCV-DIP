# Sobel, Laplacian, Canny 

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/window.jpg', 0)

# Sobel -> Sobel(image, depth, dx, dy, kSize)
sobelX = cv2.Sobel(img, -1, 1, 0) 
sobelY = cv2.Sobel(img, -1, 0, 1)
sobelXY = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)

# Laplacian -> Laplacian(image, depth)
laplacian = cv2.Laplacian(img, -1)

# Canny -> Canny(image, threshold1, threshold2)
canny = cv2.Canny(img, 80, 150)

# Prewitt 
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
kernelX = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernelY = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

prewittX = cv2.filter2D(img_gaussian,-1, kernelX)
prewittY = cv2.filter2D(img_gaussian,-1, kernelY)

plt.title("SobelX vs SobelY vs SobelXY vs Laplacian vs Canny")
result = np.hstack((kernelX, kernelY))
plt.imshow(result, cmap='gray')


```
