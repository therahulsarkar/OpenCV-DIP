# Scaling/Resizing

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/yoda.jpg')

imgScale = cv2.resize(img, None, fx=2.5, fy=2.5, interpolation=cv2.INTER_LINEAR)

plt.title("Sharp")
plt.imshow(imgScale)
```

# Roration

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/yoda.jpg')

height, width = img.shape[:2]
# height = img.shape[0]
# width = img.shape[1]

imgRotate = cv2.getRotationMatrix2D( (height/2, width/2), 90, 1) #(center, angle, scale)
imgRotate = cv2.warpAffine(img, imgRotate, (height, width)) #(img, rotateMatrix, size)

plt.title("Rotation")
plt.imshow(imgRotate)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("./img/elon.jpg")

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold the image to create a binary image
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

# Create a copy of the binary image
binary_copy = binary.copy()

# Find the contours in the binary image
contours, _ = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Define the minimum area threshold
min_area = 100

# Iterate through the contours
for contour in contours:
    # Check if the area of the contour is less than the threshold
    if cv2.contourArea(contour) < min_area:
        # Draw the contour on the binary copy with a black color (0)
        cv2.drawContours(binary_copy, [contour], 0, (0, 0, 0), -1)

# Display the original binary image
# cv2.imshow("Original Binary Image", binary)

# Display the segmented binary image
result = np.hstack((binary, binary_copy))
plt.imshow(result, cmap='gray')

```
