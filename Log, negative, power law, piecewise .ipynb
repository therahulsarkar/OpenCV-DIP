# Log transformation 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image=cv2.imread('./img/yoda.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

max_intensity = np.max(gray) #Calculating the max pixel value 

c= 255 / (np.log(1 + max_intensity)) # Calculating the value of C, c = 255/1+max
final = c * np.log(1 +image) #Calculating the image 

final = final.astype(np.uint8) #converting to 8bit uint

plt.imshow(final)
```
