# Mean filtering 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Load the image
image = cv2.imread("img/noise.png")

# Apply mean filtering
filtered_image = cv2.blur(image, (5, 5))

final_img = np.hstack((image, filtered_image));
plt.imshow(final_img)
plt.show()
```

# Median 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Load the image
image = cv2.imread("img/noise.png")

# Apply mean filtering
filtered_image = cv2.medianBlur(image, 7)

final_img = np.hstack((image, filtered_image));
plt.imshow(final_img)
plt.show()
```

# Weighted average 

```python
import cv2
import matplotlib.pyplot as plt
# Load the image
image = cv2.imread("img/noise.png")

# Define the kernel for the weighted average filter
kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]]) / 16

# Apply the weighted average filter
filtered_image = cv2.filter2D(image, -1, kernel)

final_img = np.hstack((image, filtered_image));
plt.imshow(final_img)
plt.show()
```

# Min Max

```python
import cv2
import matplotlib.pyplot as plt
# Load the image
image = cv2.imread("img/noise2.png")


# Create the kernel for the min-max filter
kernel_size = 3
kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size * kernel_size)


# Apply the weighted average filter
filtered_image = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)

plt.title("Noise image vs Image after min max filter")
final_img = np.hstack((image, filtered_image));
plt.imshow(final_img)
plt.show()
```
