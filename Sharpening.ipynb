# Using addWeighted and Gaussian

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/yoda.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gauss = cv2.GaussianBlur(img,(7,7), 2) 

op = cv2.addWeighted(img, 7.5, gauss, -6.5, 0)

plt.title("Normal vs Sharp")
result = np.hstack((img, op))
plt.imshow(result)
```
