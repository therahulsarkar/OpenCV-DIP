### For RGB 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
img_to_gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converting BGR to RGB

hist_red = cv2.calcHist([img_to_gray], [0], None, [256], [0, 256]) 
hist_green = cv2.calcHist([img_to_gray], [1], None, [256], [0, 256]) 
hist_blue = cv2.calcHist([img_to_gray], [2], None, [256], [0, 256]) 


plt.plot(hist_red)
plt.plot(hist_green)
plt.plot(hist_blue)
```

# Histogram Equilization 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
img_to_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
result = cv2.equalizeHist(img_to_gray)

input_histogram = cv2.calcHist([img], [0], None, [256], [0, 256]) 
ouput_histogram = cv2.calcHist([result], [0], None, [256], [0, 256]) 

plt.subplot(1,2,1) #plt.subplot(row, column, image_number)
plt.plot(input_histogram)

plt.subplot(1,2,2)
plt.plot(ouput_histogram)
```
