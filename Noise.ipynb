# Gaussian noise 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width, channel = img.shape

gauss_noise = np.random.normal(0, 1, img.size)
gauss_noise = gauss_noise.reshape(height, width, channel).astype(np.uint8)

final = cv2.add(img, gauss_noise)

result = np.hstack((img, final))
plt.imshow(result)
```

# Salt pepper 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

img = cv2.imread('./img/elon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
row, col, channel = img.shape

# final = img.copy()

for i in range(200000):
    x_axis = random.randint(0, col-1)
    y_axis = random.randint(0, row-1)
    
    img[y_axis, x_axis] = 255;
    
for i in range(600):
    x_axis = random.randint(0, col-1)
    y_axis = random.randint(0, row-1)
    
    img[y_axis, x_axis] = 0;
    
# result = np.hstack((img, final))
plt.imshow(img)
```
