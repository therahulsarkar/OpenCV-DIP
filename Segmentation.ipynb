# Region growing

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("./img/yoda.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


# Define the seed point for the region growing
seed_point = (100, 100)

# Perform the region growing
segmented = cv2.floodFill(img, None, seed_point, (255, 255, 255), (10, 10, 10), (10, 10, 10), cv2.FLOODFILL_FIXED_RANGE)

# Display the segmented image
plt.imshow(segmented[1], cmap='gray')
```
