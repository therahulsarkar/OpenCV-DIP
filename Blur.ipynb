# Blur

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg', 0)
mask_25 = np.ones((25,25), np.float32) / 625.0

kernel_blur = cv2.filter2D(img, -1 , mask_25)

plt.title("Normal vs Blur")
result = np.hstack((img, kernel_blur))
plt.imshow(result, cmap='gray')
```

# Blur using boxFilter 

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg', 0)

op = cv2.boxFilter(img, -1, (5,5), normalize=True) #Normalize=true means diving the mask by the multiplication 

plt.title("Normal vs Blur")
result = np.hstack((img, op))
plt.imshow(result, cmap='gray')
```

# Bilateral blur 

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg', 0)

op = cv2.bilateralFilter(img, 5, 6, 6) #image, distance, singma color, sigma space 

plt.title("Normal vs Blur")
result = np.hstack((img, op))
plt.imshow(result, cmap='gray')
```
