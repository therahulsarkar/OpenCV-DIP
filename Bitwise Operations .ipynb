# Bitwise AND, OR, NOT, XOR

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
height, width = img.shape[:2]
data = (width, height)

img2 = cv2.imread('./img/yoda.jpg')
img2_resized = cv2.resize(img2, data)

bitwise_xor = cv2.bitwise_xor(img, img2_resized)
bitwise_and = cv2.bitwise_and(img, img2_resized)
bitwise_or = cv2.bitwise_or(img, img2_resized)
bitwise_not = cv2.bitwise_not(img, img2_resized)


result = np.hstack((bitwise_xor, bitwise_and, bitwise_or, bitwise_not))
plt.imshow(result)
plt.show()
```
