# Brightness enhance 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

zero_img = np.zeros_like(img)

final = cv2.addWeighted(img, 1.5, zero_img, 0, 0)

result = np.hstack((img, final))
plt.imshow(result)
```
