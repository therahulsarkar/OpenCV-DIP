## Average of 2 images 

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('./img/birds1.jpg')
img2 = cv2.imread('./img/mountains1.jpg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img1 = img1.astype(np.float32);
img2 = img2.astype(np.float32);

img3 = (img1+img2)/2


plt.subplot(3,3,1);
plt.imshow(img1.astype(np.uint8))

plt.subplot(3,3,2)
plt.imshow(img2.astype(np.uint8))

plt.subplot(3,3,3)
plt.imshow(img3.astype(np.uint8))

plt.show()
```

## Addition 

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('./img/birds1.jpg')
img2 = cv2.imread('./img/mountains1.jpg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
# img3 = cv2.add(img1, img2)

plt.subplot(3,3,1);
plt.imshow(img1)

plt.subplot(3,3,2)
plt.imshow(img2)

plt.subplot(3,3,3)
plt.imshow(img3)

plt.show()
```

## Subtraction

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('./img/birds1.jpg')
img2 = cv2.imread('./img/mountains1.jpg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.subtract(img1, img2) #Subtracting the images 


plt.subplot(3,3,1);
plt.imshow(img1)

plt.subplot(3,3,2)
plt.imshow(img2)

plt.subplot(3,3,3)
plt.imshow(img3)

plt.show()
```

## Multiplication 

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('./img/birds1.jpg')
img2 = cv2.imread('./img/mountains1.jpg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.multiply(img1, img2)


plt.subplot(3,3,1);
plt.imshow(img1)

plt.subplot(3,3,2)
plt.imshow(img2)

plt.subplot(3,3,3)
plt.imshow(img3)

plt.show()
```

# Divide 

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('./img/birds1.jpg')
img2 = cv2.imread('./img/mountains1.jpg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.divide(img1, img2)


plt.subplot(3,3,1);
plt.imshow(img1)

plt.subplot(3,3,2)
plt.imshow(img2)

plt.subplot(3,3,3)
plt.imshow(img3)

plt.show()
```
