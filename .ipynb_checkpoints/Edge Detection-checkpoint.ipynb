# Sobel, Laplacian, Canny 

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/window.jpg', 0)

# Sobel
sobelX = cv2.Sobel(img, -1, 1, 0)
sobelY = cv2.Sobel(img, -1, 0, 1)
sobelXY = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)

# Laplacian
laplacian = cv2.Laplacian(img, -1)

# Canny
canny = cv2.Canny(img, 80, 150)

plt.title("SobelX vs SobelY vs SobelXY vs Laplacian vs Canny")
result = np.hstack((sobelX, sobelY, sobelXY, laplacian, canny))
plt.imshow(result, cmap='gray')


```

# Blur

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./img/elon.jpg', 0)
mask_25 = np.ones((25,25), np.float32) / 625.0

kernel_blur = cv2.filter2D(img, -1 , mask_25)

plt.title("SobelX vs SobelY vs SobelXY vs Laplacian vs Canny")
result = np.hstack((img, kernel_blur))
plt.imshow(result, cmap='gray')
```
